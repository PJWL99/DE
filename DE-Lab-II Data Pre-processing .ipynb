```python
dataset['Salary'].fillna(value = dataset.Salary.median(), inplace = True)
dataset['Country'].fillna(value = 'France', inplace = True)
dataset['Purchased'].fillna(value = 'Yes', inplace = True)
dataset
```

```python
# filling missing value using median()   
print ( "fill the missing values.")
median = dataset["Age"].median()
print(median)
dataset["Age"].fillna(median, inplace=True)
print(dataset.describe())
print("\n")
print(dataset)
```

```python
# drop rows with missing values
dataset.dropna(inplace=True)
dataset.shape
```

```python
# dropping duplicate values
dataset = dataset.drop_duplicates()
print(dataset)
dataset.shape
```
